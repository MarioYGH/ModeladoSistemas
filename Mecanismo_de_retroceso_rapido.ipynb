## Mecanismo de retroceso rápido


```python
#Librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
# Definición de constantes
a = 0.5
b = 1.50
c = 0.8
d = 0.6
e = 1.3
```

```python
# Definición del ángulo beta
beta = np.arange(0, 361, 0.1)
betaR = beta * np.pi / 180

# Cálculo del ángulo alfa en radianes y grados
alfaR = np.arccos(a * np.cos(betaR) / np.sqrt(a**2 + d**2 + 2 * a * d * np.sin(betaR)))
alfa = alfaR * 180 / np.pi

# Inicialización del vector xE
xE = np.zeros(len(alfaR))
x0 = 2
```

```python
# Cálculo de xE usando fsolve
for i in range(len(alfaR)):
    equation = lambda x: x**2 + 2 * x * b * np.cos(alfaR[i]) + b**2 - 2 * e * b * np.sin(alfaR[i]) + e**2 - c**2
    xE[i] = fsolve(equation, x0)[0]
    x0 = xE[i]

# Graficación
plt.plot(beta, xE, 'b', linewidth=2)
plt.xlabel('Ángulo β (grados)')
plt.ylabel('Desplazamiento bloque E (m)')
plt.xlim([min(beta), max(beta)])
plt.title('Mecanismo de retroceso rápido')
plt.show()
```
