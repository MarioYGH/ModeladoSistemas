```python
import matplotlib.pyplot as plt

# Extrayendo los valores reales de R^2 y otros errores de los archivos
r2_values = {
    "Lineal": [0.7796, 0.7818],
    "Polinomial": [0.9365, 0.9378],
    "RNN": [0.8905, 0.8945],
    "ELM": [0.9378, 0.9378]  # Basado en los resultados observados
}

mse_values = {
    "Lineal": [0.6296, 0.6320],
    "Polinomial": [0.1816, 0.1802],
    "RNN": [0.3129, 0.3055],
    "ELM": [0.1802, 0.1802]  # Basado en los resultados observados
}

mae_values = {
    "Lineal": [0.5744, 0.5779],
    "Polinomial": [0.3057, 0.3043],
    "RNN": [0.3722, 0.3629],
    "ELM": [0.3043, 0.3043]  # Basado en los resultados observados
}

times = ["Train", "Test"]

# Gráfica de R^2
plt.figure(figsize=(8, 5))
for model, values in r2_values.items():
    plt.plot(times, values, marker='o', label=f'{model}')
plt.title("Comparación de R^2 entre Modelos Predictivos")
plt.xlabel("Dataset")
plt.ylabel("R^2")
plt.legend()
plt.grid(visible=True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Gráfica de MSE
plt.figure(figsize=(8, 5))
for model, values in mse_values.items():
    plt.plot(times, values, marker='o', label=f'{model}')
plt.title("Comparación de MSE entre Modelos Predictivos")
plt.xlabel("Dataset")
plt.ylabel("MSE")
plt.legend()
plt.grid(visible=True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Gráfica de MAE
plt.figure(figsize=(8, 5))
for model, values in mae_values.items():
    plt.plot(times, values, marker='o', label=f'{model}')
plt.title("Comparación de MAE entre Modelos Predictivos")
plt.xlabel("Dataset")
plt.ylabel("MAE")
plt.legend()
plt.grid(visible=True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
